```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('students_exam_score.csv', encoding = 'unicode escape')
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['WklyStudyHours'] = df['WklyStudyHours'].str.replace('05-Oct', '5-10') 
```

```python
df.head()
```

```python
# gender distribution 
plt.figure(figsize=(6,6))
plt.title('gender distribution')
ax= sns.countplot(data = df, x = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

 from the above chart we have analyzed that the no of females in the data is more than the no of males

```python
ab1 =df.groupby('ParentMaritalStatus').agg({'MathScore':'mean' , 'ReadingScore':'mean'	, 'WritingScore':'mean'})
print(ab1)
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(ab1, annot = True)
```

from the above chart we have concluded that their is no negligible impact on the student's score due to their parent's marital status

```python
ab2 =df.groupby('NrSiblings').agg({'MathScore':'mean' , 'ReadingScore':'mean'	, 'WritingScore':'mean'})
print(ab2)
plt.figure(figsize=(6,6))
sns.heatmap(ab2, annot = True)
```

from the above chart we have concluded that their is no negligible impact on the student's score due to their no. of siblings
